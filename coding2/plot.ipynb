```python
import numpy as np 
import pandas as pd
import torch
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output.csv',header=None).values
train_acc = df[:,1]
test_acc = df[:, 3]
train_acc = [float(i[7:12]) for i in train_acc]
test_acc = [float(i[7:12]) for i in test_acc]
x = range(1,11)
train_loss = df[:,0]
test_loss = df[:,2]
plt.plot(x,train_acc)
plt.plot(x,test_acc)
plt.legend(['train_acc', 'test_acc'])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title('output_acc_without_bn')
plt.show()
plt.plot(x,train_loss)
plt.plot(x,test_loss)
plt.legend(['train_loss', 'test_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('output_loss_without_bn')
plt.show()
df
```
